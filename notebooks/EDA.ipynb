```python
# Lesson 4: Feature Engineering
```

```python
# Step 1: Load Clean Dataset
```

```python
import pandas as pd

df = pd.read_csv("cleaned_telco_churn.csv")

df.head()
```

```python
df.info()
```

```python
# Step 3: Separate Features and Target
y = df["Churn"]
```

```python
X = df.drop("Churn", axis=1)
```

```python
print(X.shape)
print(y.shape)
```

```python
# Step 4: Convert Target Variable
```

```python
y = y.map({
    "No":0,
    "Yes":1
})
```

```python
y.head()
```

```python
# Step 5: Identify Numerical Columns
numerical_features = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]
```

```python
# Step 6: Identify Categorical Columns
categorical_features = X.select_dtypes(
    include="object"
).columns

categorical_features
```

```python
# Step 7: Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Step 8: One-Hot Encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
# Step 9: Feature Scaling

StandardScaler()
```

```python
# Step 10: Build a ColumnTransformer
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

preprocessor = ColumnTransformer(

    transformers=[

        (

            "num",

            StandardScaler(),

            numerical_features

        ),

        (

            "cat",

            OneHotEncoder(handle_unknown="ignore"),

            categorical_features

        )

    ]

)
```

```python
# Step 11: Transform Data
X_train_processed = preprocessor.fit_transform(X_train)
```

```python
# Test Data
X_test_processed = preprocessor.transform(X_test)
```

```python
# Step 12: Check Shapes
```

```python
print(X_train_processed.shape)

print(X_test_processed.shape)
```

```python
# Step 13: Save Preprocessor (Recommended)
import joblib

joblib.dump(
    preprocessor,
    "preprocessor.pkl"
)
```

```python
# Next Lesson: Model Building
```

```python
# Step 1: Import Libraries
```

```python
# Machine Learning Models

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

# Evaluation

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
# Step 2: Train Logistic Regression
```

```python
lr = LogisticRegression(random_state=42)

lr.fit(X_train_processed, y_train)
```

```python
# Prediction
lr_pred = lr.predict(X_test_processed)
```

```python
# Metrics
print("Accuracy :", accuracy_score(y_test, lr_pred))
print("Precision:", precision_score(y_test, lr_pred))
print("Recall   :", recall_score(y_test, lr_pred))
print("F1 Score :", f1_score(y_test, lr_pred))
```

```python
# Step 3: Train Decision Tree
```

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train_processed, y_train)
```

```python
# Prediction
dt_pred = dt.predict(X_test_processed)
```

```python
# Meterix
print("Accuracy :", accuracy_score(y_test, dt_pred))
print("Precision:", precision_score(y_test, dt_pred))
print("Recall   :", recall_score(y_test, dt_pred))
print("F1 Score :", f1_score(y_test, dt_pred))
```

```python
# Step 4: Train Random Forest
```

```python
rf = RandomForestClassifier(
    random_state=42
)

rf.fit(
    X_train_processed,
    y_train
)
```

```python
rf_pred = rf.predict(X_test_processed)
```

```python
print("Accuracy :", accuracy_score(y_test, rf_pred))
print("Precision:", precision_score(y_test, rf_pred))
print("Recall   :", recall_score(y_test, rf_pred))
print("F1 Score :", f1_score(y_test, rf_pred))
```

```python
param_grid = {

    "n_estimators":[100,200,300],

    "max_depth":[5,10,20,None],

    "min_samples_split":[2,5,10],

    "min_samples_leaf":[1,2,4],

    "max_features":["sqrt","log2"]

}
```

```python
# Step 4: GridSearchCV
grid_search = GridSearchCV(

    estimator=RandomForestClassifier(random_state=42),

    param_grid=param_grid,

    cv=5,

    scoring="f1",

    n_jobs=-1,

    verbose=2

)
```
